```python
import pandas as pd
import matplotlib.pyplot as plt
#from google.colab import drive

#drive.mount('/content/drive')

# Load the CSV file
data = pd.read_csv('DevelopmentData.csv')

# Data cleaning and preprocessing if needed
#data['Timestamp'] = pd.to_datetime(data['Timestamp'])
print(data)
```

```python


# Assuming 'FirstObjectDistance_X' and 'FirstObjectDistance_Y' are the column names
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed

# Divide the values by 128
data['FirstObjectDistance_X'] = data['FirstObjectDistance_X'] / 128
data['FirstObjectDistance_Y'] = data['FirstObjectDistance_Y'] / 128

# Scatter plot
plt.scatter(data['FirstObjectDistance_X'], data['FirstObjectDistance_Y'], c='blue', alpha=0.5)

# Set labels and title
plt.xlabel('First Object Distance X')
plt.ylabel('First Object Distance Y')
plt.title('Scatter Plot of First Object Distance')

# Show the plot
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'FirstObjectDistance_X', 'SecondObjectDistance_X', 'ThirdObjectDistance_X', and 'FourthObjectDistance_X' are the column names
plt.figure(figsize=(15, 10))  # Adjust the figure size if needed

# Divide the values by 128 for all objects
data['FirstObjectDistance_X'] /= 128
data['FirstObjectDistance_Y'] /= 128
data['SecondObjectDistance_X'] /= 128
data['SecondObjectDistance_Y'] /= 128
data['ThirdObjectDistance_X'] /= 128
data['ThirdObjectDistance_Y'] /= 128
data['FourthObjectDistance_X'] /= 128
data['FourthObjectDistance_Y'] /= 128

# Scatter plots for each object
plt.subplot(2, 2, 1)
plt.scatter(data['FirstObjectDistance_X'], data['FirstObjectDistance_Y'], c='blue', alpha=0.5)
plt.xlabel('First Object Distance X')
plt.ylabel('First Object Distance Y')
plt.title('First Object')

plt.subplot(2, 2, 2)
plt.scatter(data['SecondObjectDistance_X'], data['SecondObjectDistance_Y'], c='green', alpha=0.5)
plt.xlabel('Second Object Distance X')
plt.ylabel('Second Object Distance Y')
plt.title('Second Object')

plt.subplot(2, 2, 3)
plt.scatter(data['ThirdObjectDistance_X'], data['ThirdObjectDistance_Y'], c='red', alpha=0.5)
plt.xlabel('Third Object Distance X')
plt.ylabel('Third Object Distance Y')
plt.title('Third Object')

plt.subplot(2, 2, 4)
plt.scatter(data['FourthObjectDistance_X'], data['FourthObjectDistance_Y'], c='purple', alpha=0.5)
plt.xlabel('Fourth Object Distance X')
plt.ylabel('Fourth Object Distance Y')
plt.title('Fourth Object')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'FirstObjectDistance_X', 'SecondObjectDistance_X', 'ThirdObjectDistance_X', and 'FourthObjectDistance_X' are the column names
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed

# Divide the values by 128 for all objects
data['FirstObjectDistance_X'] /= 128
data['FirstObjectDistance_Y'] /= 128
data['SecondObjectDistance_X'] /= 128
data['SecondObjectDistance_Y'] /= 128
data['ThirdObjectDistance_X'] /= 128
data['ThirdObjectDistance_Y'] /= 128
data['FourthObjectDistance_X'] /= 128
data['FourthObjectDistance_Y'] /= 128

# Scatter plots for all objects
plt.scatter(data['FirstObjectDistance_X'], data['FirstObjectDistance_Y'], c='blue', label='First Object', alpha=0.5)
plt.scatter(data['SecondObjectDistance_X'], data['SecondObjectDistance_Y'], c='green', label='Second Object', alpha=0.5)
plt.scatter(data['ThirdObjectDistance_X'], data['ThirdObjectDistance_Y'], c='red', label='Third Object', alpha=0.5)
plt.scatter(data['FourthObjectDistance_X'], data['FourthObjectDistance_Y'], c='purple', label='Fourth Object', alpha=0.5)

# Set labels and title
plt.xlabel('Object Distance X')
plt.ylabel('Object Distance Y')
plt.title('Scatter Plot of Object Distances')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
